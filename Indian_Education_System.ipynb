```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as sb
```

```python
Quitter= pd.read_csv("/content/Quitter-ratio-2012-2015.csv")
```

```python
num_rows, num_cols = Quitter.shape

print(f'The dataset has {num_rows} rows and {num_cols} columns.')
```

```python
Quitter.info()
```

```python
null_counts = Quitter.isnull().sum()

print("Count of null values in each column:")
print(null_counts)
```

```python
Quitter.replace('NR', '0', inplace=True)
```

```python
Quitter.describe()
```

```python
Quitter.head()
```

```python
Quitter.replace("AN", "A & N Islands", inplace=True)
```

```python
Quitter.groupby('State_UT').groups
```

```python
Quitter['Primary_Boys']=Quitter['Primary_Boys'].apply(lambda x: float(x))
```

```python
Quitter['Primary_Girls']=Quitter['Primary_Girls'].values.astype(np.float32)
```

```python
Quitter['Upper Primary_Boys'].values
```

```python
Quitter=Quitter.replace('Uppe_r_Primary',0)
```

```python
Quitter['Upper Primary_Boys']=Quitter['Upper Primary_Boys'].values.astype(np.float32)
Quitter['Upper Primary_Girls']=Quitter['Upper Primary_Girls'].values.astype(np.float32)

Quitter['Secondary _Boys']=Quitter['Secondary _Boys'].values.astype(np.float32)
Quitter['Secondary _Girls']=Quitter['Secondary _Girls'].values.astype(np.float32)

Quitter['HrSecondary_Boys']=Quitter['HrSecondary_Boys'].values.astype(np.float32)
Quitter['HrSecondary_Girls']=Quitter['HrSecondary_Girls'].values.astype(np.float32)
```

```python
Quitter.describe()
```

```python
years = Quitter['year'].unique().tolist()
print(years)
```

```python
mp.figure(figsize=(15, 8))
sb.countplot(data=Quitter, x='State_UT')
mp.xticks(rotation=90)
mp.tight_layout()
mp.show()
```

```python
from IPython.display import display

dropout_tn = Quitter[Quitter['State_UT'] == 'Tamil Nadu']
display(dropout_tn.style)

dropout_wb = Quitter[Quitter['State_UT'] == 'West Bengal']
print("Data for West Bengal:")
display(dropout_wb.style)
```

```python
mp.figure(figsize=(5, 8))
mp.pie(dropout_wb['Secondary _Girls'],labels=dropout_wb['year'], autopct='%1.1f%%')
mp.xlabel('Secondar _Girls')
mp.title('Secondary_Girls ratio based on West_Bengal')
mp.show()

mp.figure(figsize=(5, 8))
mp.pie(dropout_wb['Secondary _Boys'],labels=dropout_wb['year'], autopct='%1.1f%%')
mp.xlabel('Secondar _Boys')
mp.title('Secondary_Boys ratio based on West_Bengal')
mp.show()
```

```python
mp.figure(figsize=(5, 8))
mp.pie(dropout_wb['HrSecondary_Girls'],labels=dropout_wb['year'], autopct='%1.1f%%')
mp.xlabel('HrSecondary_Girls')
mp.title('HrSecondary_Girls ratio based on West_Bengal')
mp.show()

mp.figure(figsize=(5, 8))
mp.pie(dropout_wb['HrSecondary_Girls'],labels=dropout_wb['year'], autopct='%1.1f%%')
mp.xlabel('HrSecondary_Girls')
mp.title('HrSecondary_Girls ratio based on West_Bengal')
mp.show()
```

```python
# Outlier detection using boxplots
mp.figure(figsize= (20,15))

mp.subplot(4,4,1)
mp.title('Outliers of Primary_Girls')
sb.boxplot(Quitter['Primary_Girls'])


mp.subplot(4,4,2)
mp.title('Outliers of Primary_Boys')
sb.boxplot(Quitter['Primary_Boys'])

mp.subplot(4,4,3)
mp.title('Outliers of Upper Primary_Boys')
sb.boxplot(Quitter['Upper Primary_Boys'])

mp.subplot(4,4,4)
mp.title('Outliers of Upper Primary_Girls')
sb.boxplot(Quitter['Upper Primary_Girls'])

mp.subplot(4,4,5)
mp.title('Outliers of Secondary_Girls')
sb.boxplot(Quitter['Secondary _Girls'])

mp.subplot(4,4,6)
mp.title('Outliers of Secondary_Boys')
sb.boxplot(Quitter['Secondary _Boys'])

mp.subplot(4,4,7)
mp.title('Outliers of HrSecondary_Boys')
sb.boxplot(Quitter['HrSecondary_Boys'])

mp.subplot(4,4,8)
mp.title('Outliers of HrSecondary_Girls')
sb.boxplot(Quitter['HrSecondary_Girls'])
```

```python
Quitter['Primary_Girls']=Quitter['Primary_Girls'].clip(lower=Quitter['Primary_Girls'].quantile(0.05), upper=Quitter['Primary_Girls'].quantile(0.95))

Quitter['Primary_Boys']=Quitter['Primary_Boys'].clip(lower=Quitter['Primary_Boys'].quantile(0.05), upper=Quitter['Primary_Boys'].quantile(0.95))

Quitter['Upper Primary_Boys']=Quitter['Upper Primary_Boys'].clip(lower=Quitter['Upper Primary_Boys'].quantile(0.05), upper=Quitter['Upper Primary_Boys'].quantile(0.95))

Quitter['Upper Primary_Girls']=Quitter['Upper Primary_Girls'].clip(lower=Quitter['Upper Primary_Girls'].quantile(0.05), upper=Quitter['Upper Primary_Girls'].quantile(0.95))

Quitter['Secondary _Girls']=Quitter['Secondary _Girls'].clip(lower=Quitter['Secondary _Girls'].quantile(0.05), upper=Quitter['Secondary _Girls'].quantile(0.95))

Quitter['Secondary _Boys']=Quitter['Secondary _Boys'].clip(lower=Quitter['Secondary _Boys'].quantile(0.05), upper=Quitter['Secondary _Boys'].quantile(0.95))

Quitter['HrSecondary_Boys']=Quitter['HrSecondary_Boys'].clip(lower=Quitter['HrSecondary_Boys'].quantile(0.05), upper=Quitter['HrSecondary_Boys'].quantile(0.95))

Quitter['HrSecondary_Girls']=Quitter['HrSecondary_Girls'].clip(lower=Quitter['HrSecondary_Girls'].quantile(0.05), upper=Quitter['HrSecondary_Girls'].quantile(0.95))

```

```python
# Outlier detection using boxplots
mp.figure(figsize= (20,15))

mp.subplot(4,4,1)
mp.title('Outliers of Primary_Girls')
sb.boxplot(Quitter['Primary_Girls'])


mp.subplot(4,4,2)
mp.title('Outliers of Primary_Boys')
sb.boxplot(Quitter['Primary_Boys'])

mp.subplot(4,4,3)
mp.title('Outliers of Upper Primary_Boys')
sb.boxplot(Quitter['Upper Primary_Boys'])

mp.subplot(4,4,4)
mp.title('Outliers of Upper Primary_Girls')
sb.boxplot(Quitter['Upper Primary_Girls'])

mp.subplot(4,4,5)
mp.title('Outliers of Secondary_Girls')
sb.boxplot(Quitter['Secondary _Girls'])

mp.subplot(4,4,6)
mp.title('Outliers of Secondary_Boys')
sb.boxplot(Quitter['Secondary _Boys'])

mp.subplot(4,4,7)
mp.title('Outliers of HrSecondary_Boys')
sb.boxplot(Quitter['HrSecondary_Boys'])

mp.subplot(4,4,8)
mp.title('Outliers of HrSecondary_Girls')
sb.boxplot(Quitter['HrSecondary_Girls'])
```

```python
df1 = pd.read_csv('/content/schools-with-boys-toilet-2013-2016.csv')
df2 = pd.read_csv('/content/schools-with-girls-toilet-2013-2016.csv')

Dataset = pd.concat([df1, df2])
Dataset.set_index('State_UT', inplace=True)

print(Dataset)
display(Dataset)
```

```python
Dataset.info()
```

```python
Dataset.replace('NR','0',inplace = True)
```

```python
Dataset.columns
```

```python
Dataset.isnull().sum()
```

```python
mp.figure(figsize=(10, 6))

width = 0.4

mp.bar(Dataset.index, Dataset['All Schools'], width=width, align='center', label='All Schools')

# Set labels and title
mp.xlabel('States')
mp.xticks(rotation=90)
mp.ylabel('Count')
mp.title('Number of Schools based on states')

# Display the plot
mp.show()
```

```python
mp.figure(figsize=(10, 6))
width = 0.4
mp.bar(Dataset['All Schools'], Dataset['year'])
mp.title('Number of Schools based on Year')
mp.show()
```

```python
Dataset['Primary_Only'] = Dataset['Primary_Only'].values.astype(np.float32)
Dataset['U_Primary_Only'] = Dataset['U_Primary_Only'].values.astype(np.float32)
Dataset['Sec_Only'] = Dataset['Sec_Only'].values.astype(np.float32)
Dataset['HrSec_Only'] = Dataset['HrSec_Only'].values.astype(np.float32)
```

```python
# Outlier detection using boxplots
mp.figure(figsize= (20,15))
mp.subplot(4,4,1)
sb.boxplot(Dataset['Primary_Only'])

mp.subplot(4,4,2)
sb.boxplot(Dataset['U_Primary_Only'])

mp.subplot(4,4,3)
sb.boxplot(Dataset['Sec_Only'])

mp.subplot(4,4,4)
sb.boxplot(Dataset['HrSec_Only'])
```

```python
Dataset['Primary_Only'] = Dataset['Primary_Only'].clip(lower=Dataset['Primary_Only'].quantile(0.05), upper=Dataset['Primary_Only'].quantile(0.95))

Dataset['U_Primary_Only'] = Dataset['U_Primary_Only'].clip(lower=Dataset['U_Primary_Only'].quantile(0.05), upper=Dataset['U_Primary_Only'].quantile(0.95))
```
